**1) What are the mean carat and the median carat for the each group of color in this dataset?** 

```python
# Code here to find the mean carat and median carat
df.groupby('color')['carat'].mean().reset_index()
```

```python
df.groupby('color')['carat'].median().reset_index()
```

```python
df.groupby('color').agg({'carat':['mean','median']})
```

**BONUS 1: Use the seaborn library to make a scatter plot between carat and depth, carat and table, carat and price. Use the hue argument to group by either color or cut**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12, 12))

plt.subplot(3, 1, 1)
sns.scatterplot(x="carat", y="depth", data=df, hue='color')

plt.subplot(3, 1, 2)
sns.scatterplot(x="carat", y="table", data=df, hue ='cut')

plt.subplot(3, 1, 3)
sns.scatterplot(x="carat", y="price", data=df, hue ='cut')

plt.show()
```
